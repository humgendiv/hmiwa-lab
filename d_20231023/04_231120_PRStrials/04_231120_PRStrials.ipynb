File headers: 
1. snpname - dbSNP rsID 
2. chr - chromosome 
3. pos - position 
4. markername - chr:pos 
5. ref - reference allele (hg19 + strand) 
6. alt - alternate allele (hg19 + strand) 
7. beta - beta 
8. se - standard error 
9. pvalue - P value 
10. n - sample size 
11. gmaf/eur_maf - alternate allele frequency in 1000 Genome Combined/European Ancestries 
12. exac_maf/exac_nfe_maf -alternate allele frequency in ExAC Combined/Non-Finnish European Ancestries

```bash
#起動
conda activate
conda activate ldsc_env
```

```bash
#面倒なので呪文を設定
ldscdir=~/hmiwa/hmiwa-lab/d_20231023/04_231120_PRStrials/ldsc
```

```bash
head BMI_All_ancestry.txt
```

```bash
python $ldscdir/munge_sumstats.py \
--sumstats BMI_All_ancestry.txt \
--out BMI_All_ancestry.out \
--snp SNPNAME \
--a1 REF \
--a2 ALT \
--p Pvalue \
--N 526508
```
